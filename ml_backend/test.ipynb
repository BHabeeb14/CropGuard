```python
import tensorflow as tf
import os

# 1. Use the path where kagglehub extracted the files
# Note: Use double backslashes or a raw string for Windows paths
data_dir = r"C:\Users\taotechs\.cache\kagglehub\datasets\abdallahalidev\plantvillage-dataset\versions\3"

# 2. Check the subfolders to ensure the data is there
# Your console error showed 0 files before; this ensures you are in the right subfolder
print("Subfolders found:", os.listdir(data_dir))

# 3. Load Training Dataset (70%)
train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.3,
    subset="training",
    seed=123,
    image_size=(224, 224),
    batch_size=32
)

# 4. Load Validation/Test Dataset (30%)
val_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.3,
    subset="validation",
    seed=123,
    image_size=(224, 224),
    batch_size=32
)
```

```python
import matplotlib.pyplot as plt

# 1. Take one batch of 32 images from the training set
plt.figure(figsize=(15, 10))
for images, labels in train_ds.take(1):
    for i in range(9): # View the first 9 images
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(train_ds.class_names[labels[i]])
        plt.axis("off")
```
